```
from pyspark.sql import SparkSession
from pyspark.sql.functions import *
```

```
spark=SparkSession.builder.appName("CSVPractice").getOrCreate()
```

```
df=spark.read.csv("/content/pyspark_large_dummy_dataset.csv",
header=True,
inferSchema=True)
```

```
df.show(5)
```

```
df.printSchema()
```

```
df.select("Customer_Name", "Price").show()
```

```
df.filter(df.Price>50000).show(40)
```

```
df.filter(df.Order_ID.between(10,35)).show(40)
```

```
df.filter(df.Price.between(20000,50000)).show(100)
```

```
df.filter(df.Price==30000).show()
```

```
#df.where() is identical to df.filter()
df.where(df.Order_Date=="2025-05-23").show()
```

```
df.withColumn("Total", col("Quantity")*col("Price")).show()
```

```
df.show(5)
df.count()
```

```
len(df.columns)
```

```
df.show(5)
```

```
df.groupBy("product") \
.agg(avg("Price"),max("Price"),min("Price"),count("Product")) \
.show()
```

```
df.groupBy("Customer_Name") \
.agg(collect_list("Price")).show()
```

```
df.groupBy("Customer_Name") \
.agg(sum("Price").alias("Total_Price")).show()
```

```
df.orderBy(col("Price").desc()).show()
```

```
df.withColumn("Total", col("Quantity")*col("Price")) \
.orderBy(col("Total").desc()) \
.show()
```

```
df.describe().show()
```

```
df.sort("Price").show()
```

```
df.distinct().show()
```

```
df.select("Customer_Name").distinct().show()
```
