```
from pyspark.sql import SparkSession
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```
spark=SparkSession.builder.appName("Delhi_Air_Quality").getOrCreate()
```

```
df=spark.read.csv("/content/Delhi Air Quality Time Series Dataset.csv", inferSchema=True, header=True)
```

```
df.schema
```

```
df.printSchema()
```

```
df.show()
```

```
df.count()
```

```
len(df.columns)
```

```
df.select("Values", "DateTime","hour").dtypes
```

```
df.select("Locations","Parameters","Values").show()
```

```
df.select("Locations", "Parameters",(col("Values")>300)).show()
```

```
df.filter(col("Values")>300).show()
```

```
df.select("Parameters").distinct().show()
```

```
df.orderBy(col("Values").desc()).show(10)
```

```
#night-time pollution
df.filter(col("hour")>20).show()
```

```
df.select(col("Locations")).distinct().show()
```

```
df.groupBy("Locations").agg(avg("Values")).show()
```

```
df.groupBy("Parameters").agg(max("Values")).show()
```

```
df.groupBy("Locations").agg(count("Parameters")).show()
```

```
df = df.withColumn("Pollution_Level",
                   when(col("Values")>300, "High")
                   .when((col("Values")>100) & (col("Values")<300), "Medium")
                   .otherwise("Low"))
```

```
df.show()
```

```
df.printSchema()
```

```
df.groupBy("Locations").agg(max("Values").alias("Highest_Pollution")).show()
```

```
# df.agg(max("Values")).show()
df.orderBy(col("Values").desc()).show(1)
```

```
df.groupBy(col("hour")).agg(avg("Values").alias("Avg_Pollution")).sort("hour").show(25)
```

```
df = df.withColumnRenamed("Values", "Pollution_Value")
```

```
df.columns
```

```
df.show()
```

```
df = df.dropDuplicates()
```

```
df.show()
```

```
df.describe("Pollution_Value").show()
```

```
df.agg(min("Pollution_Value")).show()
```

```
df.groupBy(col("Locations")).agg(min("Pollution_Value")).show()
```

```
df.where(col("Parameters") == "pm25").show()
```

```
new_df=df.filter(col("Pollution_Level") == "High")
```

```
df.show()
new_df.show()
```

```
new_df.coalesce(1).write.mode("overwrite").csv("/content/High_Dataset",header=True)
```

```
dashboard_df=df.groupBy(col("Locations")).agg(avg("Pollution_Value"),max("Pollution_Value"),count("*").alias("Total_Records"))
```
